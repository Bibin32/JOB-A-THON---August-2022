# X and y

```python
req_cols =['campaign_id','click_rate',]
X = df.drop(['click_rate','campaign_id','is_timer','body_len','subject_len','target_audience','mean_CTA_len'], axis =1)
y = pd.DataFrame(df['click_rate'])
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,train_size=0.2,random_state=42)
```

```python
X_test.shape
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
```

```python
model = lr.fit(X_train,y_train)
```

```python
lr_predict = model.predict(X_test)
```

```python
lr_predict
```

```python
from sklearn.metrics import mean_squared_error, r2_score
```

```python
MSE = mean_squared_error(y_test,lr_predict)
```

```python
MSE
```

```python
r2_score(y_test,lr_predict)
```

# KNN

```python
# Import the knn regressor class 
from sklearn.neighbors import KNeighborsRegressor
# Create an instance of the class
knn = KNeighborsRegressor(n_neighbors=146)
```

```python
knn.fit(X_train,y_train)
```

```python
y_pred_knn = knn.predict(X_test)
```

```python
mean_squared_error(y_test,y_pred_knn)
```

# Decision Tree Regressor

```python
from sklearn.tree import DecisionTreeRegressor
# Create an instance of the class
dt_reg = DecisionTreeRegressor(random_state=45,max_depth=88)
```

```python
dt_reg.fit(X_train,y_train)
```

```python
y_pred_dt = dt_reg.predict(X_test)
```

```python
mean_squared_error(y_test,y_pred_dt)
```

```python
r2_score(y_test,y_pred_dt)
```

# Bagging Regressor

```python
from sklearn.ensemble import BaggingRegressor
dt_reg = DecisionTreeRegressor(max_depth=16)
bg_regr = BaggingRegressor(base_estimator=dt_reg, n_estimators=500, random_state=123)
```

```python
bg_regr.fit(X_train,y_train)
```

```python
y_pred_bg = bg_regr.predict(X_test)
```

```python
mean_squared_error(y_test,y_pred_bg)
```

```python
r2_score(y_test,y_pred_bg)
```
